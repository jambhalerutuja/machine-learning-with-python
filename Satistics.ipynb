```python
sns.countplot(x="Sex",data=df)
```

```python
sns.relplot(x="Age",y="Pclass",data=df)
```

```python
sns.relplot(x="Age",y="Pclass",hue="Sex",data=df)
```

```python
sns.catplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
sns.catplot(x="Embarked",y="Age",hue="Sex",jitter=False,data=df)
```

```python
sns.catplot(x="Embarked",y="Age",hue="Sex",kind="swarm",data=df)
```

```python
sns.boxplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
sns.barplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
sns.pointplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
sns.distplot(df.Age)
```

```python
sns.jointplot(x="Age",y="Pclass",hue="Sex",data=df)
```

```python
sns.jointplot(x=df.Age,y=df.Pclass,kind="hex",data=df)
```

```python
sns.jointplot(x=df.Age,y=df.Pclass,kind="kde",data=df)
```

```python
corr=df.corr()
plt.figure(figsize=(10,12))
f,ax=plt.subplots()
sns.heatmap(corr,vmax=.8,square=True)
```
