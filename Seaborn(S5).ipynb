```python
sns.relplot(x="Age",y="Fare",data=df)
```

```python
sns.relplot(x="Age",y="Fare",hue="Sex",data=df)
```

```python
 sns.catplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
 sns.catplot(x="Embarked",y="Age",hue="Sex",jitter=False,data=df)
```

```python
 sns.catplot(x="Embarked",y="Age",kind = "swarm",data=df)
```

```python
 sns.catplot(x="Embarked",y="Age",hue="Sex",kind = "swarm",data=df)
```

```python
 sns.boxplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
 sns.barplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
 sns.pointplot(x="Embarked",y="Age",hue="Sex",data=df)
```

```python
sns.distplot(df.Age)
```

```python
sns.distplot(df.Age,kde=False,rug=True)
```

```python
sns.jointplot(x="Age",y="Fare",data=df)
```

```python
sns.jointplot(x=df.Age,y=df.Fare,kind="hex",data=df)
```

```python
sns.jointplot(x="Age",y="Fare",kind="kde",data=df)
```

```python
corr=df.corr()
plt.figure(figsize=(10,10))
f, ax = plt.subplots()
sns.heatmap(corr,vmax=.8,square=True)
```
