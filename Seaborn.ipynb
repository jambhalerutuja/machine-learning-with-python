```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/Premalatha-success/Datasets/main/titanic-training-data.csv")
```

```python
sns.relplot(x="Age",y="Fare",data=df)
```

```python
sns.relplot(x="Age",y="Fare",hue="Sex",data=df)
```

```python
sns.catplot(x="Survived",y="Fare",hue="Sex",data=df)
```

```python
sns.catplot(x="Sex",y="Age",kind="swarm",data=df)
```

```python
sns.relplot(x="Age",y="Fare",size="Survived",hue="Sex",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex",data=df)
```

```python
sns.catplot(x="Pclass",y="Age",hue="Sex",jitter=False,data=df)
```

```python
sns.countplot(x="Sex",data=df)
```
