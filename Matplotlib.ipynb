```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpoints=np.array([1,2])
ypoints=np.array([5,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,20])
ypoints=np.array([10,200])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints=np.array([1,6])
ypoints=np.array([4,5])
plt.plot(xpoints,ypoints,"*")
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,"X" 'g')
plt.show()
```

```python
ypoints=np.array([3,8,1,12])
xpoints=np.array([5,6,7,12])
plt.plot(xpoints,ypoints,marker='*',ms=20,color='blue')
plt.show()
```

```python
ypoints=np.array([3,8,1])
plt.plot(ypoints,marker='o',ms=50,mec='black',color='pink')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o',ms=30,mfc='purple',mec='white')
plt.show()
```

```python
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,linestyle='dashed',marker='o',mec='red',ms=20,mfc='pink',linewidth=1.5)
plt.show()
```

```python
xpoints=np.array([10,20,30])
ypoints=np.array([1,2,6])
plt.plot(xpoints,ypoints,marker='*',color='blue',mfc='black',mec='pink',ms=40,linestyle='dotted')
plt.show()
```

```python
y1=np.array([3,8,1,10])
y2=np.array([6,2,5,11,10,12])
plt.plot(y1,color='g',marker='*')
plt.plot(y2,color='b',marker='.')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125])
y=np.array([240,250,260,270,280,290,300,310,320,330])
plt.plot(x,y,color='blue',linewidth=1)
plt.xlabel('avg pulse',fontsize=10,color='brown')
plt.ylabel('calorie burnage',fontsize=10,color='brown')
plt.title('SAMPLE',fontsize=25,color='green')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125,144])
y=np.array([220,240,260,270,280,290,310,315,320,330,340])
plt.plot(x,y,color='red',linewidth=1)
plt.xlabel('avg pulse',fontsize=10,color='brown')
plt.ylabel('calorie burnage',fontsize=10,color='brown')
plt.title('SAMPLE',fontsize=25,color='blue')
plt.grid(color='green',linestyle='dotted')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125,144])
y=np.array([220,240,260,270,280,290,310,315,320,330,340])
plt.plot(x,y,color='red',linewidth=1)
plt.xlabel('avg pulse',fontsize=10,color='brown')
plt.ylabel('calorie burnage',fontsize=10,color='brown')
plt.title('SAMPLE',fontsize=25,color='blue')
plt.grid(axis='x')
plt.show()
```

```python
x=np.array([80,85,90,95,100,105,110,115,120,125,144])
y=np.array([220,240,260,270,280,290,310,315,320,330,340])
plt.plot(x,y,color='red',linewidth=1)
plt.xlabel('avg pulse',fontsize=10,color='brown')
plt.ylabel('calorie burnage',fontsize=10,color='brown')
plt.title('SAMPLE',fontsize=25,color='blue')
plt.grid(axis='y')
plt.show()
```

```python
y=np.array([10,70,44,52,6])
mylabels=['Apple','mango','berry','melon','orange']
plt.pie(y,labels=mylabels)
plt.show()
```

```python
y=np.array([2,10,4,7,6])
mylabels=['Apple','mango','berry','melon','orange']
myexplode=[0,0.15,0.4,0.5,0.6]
plt.pie(y,labels=mylabels, explode=myexplode)
plt.show()
```

```python
y=np.array([10,70,44,52,6])
mylabels=['Apple','mango','berry','melon','orange']
plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```
